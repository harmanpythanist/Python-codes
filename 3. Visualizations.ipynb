```python
year = ['2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025', '2026']

baber_scores = [900, 1400, 800, 900, 1250, 2000, 500, 800, 1500]
kohli_scores = [500, 1200, 400, 900, 1300, 1500, 2100, 1500, 2500]
maxwell_scores = [1200, 1500, 1100, 2500, 2800, 3100, 1800, 3500, 4000]
```

```python
plt.plot(year, baber_scores, label = 'BABER AZAM')
plt.scatter(year, baber_scores)

plt.plot(year, kohli_scores, label = 'VIRAT KOHLI')
plt.scatter(year, kohli_scores)

plt.plot(year, maxwell_scores, label = 'Maxwell')
plt.scatter(year, maxwell_scores)

plt.legend()
plt.show()
```

```python
year = ['2020', '2021', '2022', '2023', '2024', '2025']

pak_population = [100, 130, 160, 170, 210, 250]
india_population = [300, 350, 460, 500, 540, 600]
```

```python
plt.plot(year, pak_population, label = 'PAKISTAN')
plt.scatter(year, pak_population)

plt.plot(year, india_population, label = 'INDIA')
plt.scatter(year, india_population)

plt.title('POPULATION IN MILLIONS')
plt.ylabel('POPULATION IN MILLiONS')
plt.xlabel('YEARS')
plt.legend()
plt.show()
```
