```python
random_numbers = np.linspace(1,10,100)
```

```python
refined = np.cos(random_numbers)
```

```python
plt.plot(random_numbers, refined)
plt.title('Graph of COS function')
```

```python
refined = np.sin(random_numbers)
plt.scatter(random_numbers, refined)
plt.title('Graph of SIN function')
```

```python
refined = np.tan(random_numbers)
plt.plot(random_numbers, refined)
plt.title('Graph of TAN function')
```
